```python
import pandas as pd
import numpy as np
```

```python
# Leitura da base 
df = pd.read_csv("./data.csv", sep=",", header=0, encoding="latin1")
df.head()
```

```python
# visualizando as primeiras informações
df.info
```

```python
# Removendo valores nulos da coluna "Customer ID"
df = df.dropna(subset=['CustomerID'])
```

```python
# Convertendo a data de texto para formato de data real
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# Criando coluna de Faturamento (TotalLine)
df['TotalLine'] = df['Quantity'] * df['UnitPrice']
```

```python
# Criando coluna "Status" para ficar mais facil para fazer o dashboard depois
df['Status'] = df['InvoiceNo'].apply(lambda x: 'Cancelado' if str(x).startswith('C') else 'Venda')
```

```python
# Salvando arquivo tratado
df.to_csv('data_limpo.csv', index=False)

print("\n--- Limpeza Concluída! ---")
print(f"Linhas restantes: {len(df)}")
print(df.head())
```

```python
# importei pro PostgreSQL e deu erro porque o customer ID esta como float e nao como int
# então vou fazer a correção por aqui e transformar em int
df['CustomerID'] = df['CustomerID'].astype(int)
```

```python
# Salvando novamente
df.to_csv('ecommerce_limpo.csv', index=False)

print("\n--- Limpeza Concluída! ---")
print(f"Linhas restantes: {len(df)}")
print(df.head())
```
